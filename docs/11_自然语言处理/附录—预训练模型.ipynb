## 6. 命名实体识别
命名实体识别（Named Entity Recognition，NER）是数据预处理中的一个可选步骤，旨在将文本中的命名实体定位并分类到预定义的类别中。这些类别可以包括人名、组织名、地点名、时间表达式、数量、货币值或百分比等。

```python
from transformers import pipeline

# 加载预训练的NER模型
nlp = pipeline("ner", grouped_entities=True)

# 示例文本
text = "Apple is looking at buying U.K. startup for $1 billion."

# 识别命名实体
entities = nlp(text)
for entity in entities:
    print(entity)
```
